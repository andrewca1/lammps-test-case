```python
%matplotlib inline

from __future__ import print_function
from statsmodels.compat import lzip
import statsmodels
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
from statsmodels.formula.api import ols
from statsmodels.graphics.api import interaction_plot, abline_plot
from statsmodels.stats.anova import anova_lm
import matplotlib.pyplot as plt

# Change font size
import matplotlib
matplotlib.rcParams.update({'font.size': 16})
```

```python
# Load data
dat = pd.read_csv("uq.csv")

# Throw columns into variables
DA, DB, DC, k, cC = dat.DA, dat.DB, dat.DC, dat.k, dat.cC
```

```python
plt.figure(figsize=(20,20))
plt.subplot(221)
plt.scatter(DA, cC, s=100)
plt.xlim([min(DA), max(DA)])
plt.ylim([min(cC), max(cC)])
plt.xlabel('DA')
plt.ylabel('cC')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))

plt.subplot(222)
plt.scatter(DB, cC, s=100)
plt.xlim([min(DB), max(DB)])
plt.ylim([min(cC), max(cC)])
plt.xlabel('DB')
plt.ylabel('cC')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))

plt.subplot(223)
plt.scatter(DC, cC, s=100)
plt.xlim([min(DC), max(DC)])
plt.ylim([min(cC), max(cC)])
plt.xlabel('DC')
plt.ylabel('cC')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))

plt.subplot(224)
plt.scatter(k, cC, s=100)
plt.xlim([min(k), max(k)])
plt.ylim([min(cC), max(cC)])
plt.xlabel('k')
plt.ylabel('cC')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0));
```

```python
formula = 'cC ~ DA + DB + DC + k'
lm = ols(formula, dat).fit()
print(lm.summary())
```
